## Zadanie 1 - Suma sygnałów

```julia
using Plots, FFTW
```

```julia
Fs = 1024
t = 0:1/(Fs-1):1;
x = sin.(2*pi*t*200) + 2* sin.(2*pi*t*400)
plot(t,x)
```

```julia
y = fft(x)
sticks((abs.(y)))
```

## Zadanie 2 - usuwanie szumów

```julia
Fs = 1024
t = 0:1/(Fs-1):1
szumy = cos.(2*pi*t*200) .+ rand(length(t))
plot(t,szumy)
```

```julia
y = fft(szumy)
sticks((abs.(y)))
```

```julia
widmo = abs.(y)
C = findall(widmo.<=50)
widmo[C] .= 0


filter = copy(y)
for i in 1:length(filter)
    if abs(filter[i])
        filter[i] = 0
```

```julia
plot(real(ifft(widmo)))
```
